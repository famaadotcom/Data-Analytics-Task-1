```python
# ==========================================
# Task 1: Exploring and Visualizing Iris Dataset
# ==========================================

# 1. Import Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import files

# 2. Upload and Load file
uploaded = files.upload()

df = pd.read_csv('Iris.csv')

# 3. Dataset Structure (As per Instructions)
print("\n--- DATASET SHAPE ---")
print(df.shape)

print("\n--- COLUMN NAMES ---")
print(df.columns)

print("\n--- FIRST 5 ROWS ---")
print(df.head())

# 4. Visualizations
sns.set_theme(style="whitegrid")

# (A) Scatter Plot
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='SepalLengthCm', y='SepalWidthCm', hue='Species')
plt.title('Scatter Plot: Relationship between Sepal Length & Width')
plt.show()

# (B) Histogram
plt.figure(figsize=(8, 5))
sns.histplot(df['PetalLengthCm'], kde=True, color='purple')
plt.title('Histogram: Distribution of Petal Length')
plt.show()

# (C) Box Plot
plt.figure(figsize=(8, 5))
sns.boxplot(x='Species', y='PetalWidthCm', data=df)
plt.title('Box Plot: Spread of Petal Width by Species')
plt.show()
```
